```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Anna's results
keywords = [
    'model', 'use', 'training', 'language', 'task', 'train', 'image', 'network', 'dataset', 'result']
frequencies = [
    78186, 42882, 28528, 27349, 23502, 20948, 19266, 19247, 18421, 18229]
```

```python
# Generating simulated data for keywords over time
np.random.seed(0)  # For reproducibility
years = np.arange(2018, 2023)  # 5-year period
word_trends = {word: frequencies[i] * (1 + (np.random.rand(len(years)) - 0.5) / 10) for i, word in enumerate(keywords)}
```

```python
# set Times new roman
plt.rcParams['font.family'] = 'Times New Roman'

# Creating the subplot
fig, axs = plt.subplots(1, 2, figsize=(8, 3))

# First plot: Bar plot of keywords frequencies
axs[0].barh(keywords, frequencies, color='skyblue')
axs[0].set_title('Top 10 Keywords Frequency')
axs[0].set_ylabel('Frequency')
axs[0].tick_params(axis='x', rotation=45)

# Second plot: Line plot of each word frequency over a 5-year period
for word, trend in word_trends.items():
    axs[1].plot(years, trend, label=word)
axs[1].set_title('Keywords Frequency Trend Over 5 Years')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Frequency')
axs[1].legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()

# Save the plot
plt.savefig('../figures/keyword_frequencies.png', dpi=100, bbox_inches='tight')
plt.show()
```
